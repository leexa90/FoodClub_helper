```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 25)
import statsmodels.api as sm
import statsmodels.formula.api as smf
df0 = pd.read_csv('./data_3575-6015.csv')
df0['test'] = 0

df=pd.concat([df0,]).reset_index(drop=1)
df['arena'] = sorted([0,1,2,3,4]*4)*(len(df)//20)

odds = []
for i in range(0,len(df),4):
    odds += [list(df.iloc[i:i+4]['open_odds'].values),]*4
df['odds_type'] = odds
df['odds_type'] = df['odds_type'].astype(str)
df['win'] = df['win/lose'].map({'Loser':0,'Winner':1})
df['open_odds_norm'] = 1/df['open_odds']/df['sum_(1/open_odds)']
df['close_odds_norm'] = 1/df['close_odds']/df['sum_(1/close_odds)']
df['constant']=1
#df['open_odds_norm2']=df['open_odds_norm']**2
df['change']=np.clip(df['close_odds_norm']-df['open_odds_norm'],-1,1)
var = 'close'
arena_threshold=1.03 #minimum expectation to consider the bet


y = df['win']

params = {2 : 0.000,
          3 : 0.292,
          4 : 0.225,
          5 : 0.183,
          6 : 0.155,
          7 : 0.134,
          8 : 0.118,
          9 : 0.106,
          10: 0.095,
          11: 0.091,
          12: 0.080,
          13: 0.050}

for i in range(0,len(df),4):
    temp = df['open_odds'].iloc[i:i+4]
    min_ = temp == temp.min()
    temp2 = temp.map(params)
    leftover_prob = 1-temp2[min_==False].sum()
    temp2.loc[min_] = leftover_prob/sum(min_)
    df.loc[i:i+3,'pred'] = temp2.values
    df.loc[i:i+3,'min_odds_inArena'] = min_.values
    
df_ori = df.copy()

print(df_ori['pred'])



```

```python
for max_payoff in [85,]:
    alpha=0.08 
    df = df_ori.sort_values(['game','arena']).reset_index(drop=1).copy()
    for i in range(0,len(df),4):
        df.loc[i:i+3,'pred']=df.loc[i:i+3,'pred']/np.sum(df.loc[i:i+3,'pred'])
    df['Expected_profit_at_close'] = df['pred']*df['close_odds']
    df['Expected_profit_at_open'] = df['pred']*df['open_odds']

    df['play'] = 0
    df=df.iloc[:].reset_index(drop=1)

    var_used = 'Expected_profit_at_'+var

    dictt_winnings = {}
    df_all_combs=pd.DataFrame([0,1,2,3,4],columns=[0,])
    for i in range(1,5):
        temp = df_all_combs.copy()
        for val in range(5):
            temp[i] = val
            df_all_combs = df_all_combs.append(temp)
        df_all_combs  = df_all_combs.dropna().astype(np.int32)

    from multiprocessing import Pool
    df = df.sort_values(['game','arena','open_odds']).reset_index(drop=1)


    def f(i,df=df,df_all_combs=df_all_combs,dictt_winnings=dictt_winnings):
        if i %10000==0:
            print (i)
        game1 = df.iloc[i:i+4]
        game2 = df.iloc[i+4:i+8]
        game3 = df.iloc[i+8:i+12]
        game4 = df.iloc[i+12:i+16]
        game5 = df.iloc[i+16:i+20]
        dictt={}
        counter =0 

        for game in [game1,game2,game3,game4,game5]:
            dictt[counter] = [np.concatenate([game['pred'].values,np.array([1,])]), #predictions
                              np.concatenate([game[var+'_odds'].values,np.array([1,])]), #closing odds (predicted)
                              np.concatenate([game['close_odds'].values,np.array([1,])]), #closing odds (actual)
                              np.concatenate([game['win'].values,np.array([1,])])] #correct choice, choosing no priates also correct
            if np.max(game[var_used]>arena_threshold)==1:

                play_id = game[game[var_used]==np.max(game[var_used])].index
                df.loc[play_id,'play'] = 1
            counter += 1

        temp = df_all_combs.copy()
        for j in range(5):
            temp['prob_'+str(j)] =temp[j].apply(lambda x : dictt[j][0][x])
            temp['odds_p_'+str(j)] =temp[j].apply(lambda x : dictt[j][1][x])
            temp['odds_a_'+str(j)] =temp[j].apply(lambda x : dictt[j][2][x])
            temp['win_'+str(j)] =temp[j].apply(lambda x : dictt[j][3][x])
        temp['correct']=    temp['win_0']*temp['win_1']* temp['win_2']* temp['win_3']* temp['win_4']    
        temp['payoff_p'] = temp['odds_p_0']*temp['odds_p_1']* temp['odds_p_2']* temp['odds_p_3']* temp['odds_p_4']
        temp['payoff_a'] = temp['odds_a_0']*temp['odds_a_1']* temp['odds_a_2']* temp['odds_a_3']* temp['odds_a_4']
        temp['prob'] =  temp['prob_0']*temp['prob_1']* temp['prob_2']* temp['prob_3']* temp['prob_4']
        temp['Expected'] = temp['payoff_p']*temp['prob'] #predicted earnings
        temp['Actual']= temp['payoff_a']*temp['correct'] #actual earnings
        temp = temp[(temp['payoff_p']<max_payoff)&(temp['Expected']>arena_threshold)].sort_values('Expected').reset_index(drop=1)
        if len(temp)==0:
            return (-0.001,0)
        threshold = temp.iloc[max(-10,-len(temp))]['Expected']
        temp2 = temp[temp['Expected'] >= threshold -alpha].reset_index(drop=1) #you bet on these 10 if >10
        temp2['chosen'] = 0
        if len(temp2)==0:
            return (-0.001,0)
        else:
            while np.sum(temp2['chosen']) <10 and np.mean(temp2['chosen']) < 1:
              #iteratively choose bet 
              max_prob = temp2[temp2['chosen']==0]['Expected'].max()
              candidates = temp2[(temp2['chosen']==0) & (temp2['Expected']>=max_prob-alpha)]
              max_prob = candidates['prob'].max()
              chosen = candidates[candidates['prob']==max_prob ].index
              temp2.loc[chosen,'chosen']=1
            temp2 =temp2[temp2['chosen']==1].reset_index(drop=1)
            dictt_winnings[game3['game'].values[0]] = np.sum(temp2['Actual'])
            temp2 = temp2.sort_values('Expected',ascending=False).iloc[:10] #got some bug, gota remake <=10
            return (np.sum(temp2['Actual']),len(temp2))

        #print(dictt_winnings[game3['game'].values[0]])     


    p = Pool(10)
    winnings_list = list(p.map(f,range(0,len(df),20)))
    #print(winnings_list)

    winnings = np.sum([x[0] for x in winnings_list if x[1]>0])
    net_profit = [x[0]-x[1] for x in winnings_list if x[1]>0]
    list_winnings = [x[0] for x in winnings_list if x[1]>0]
    winnings_stE = np.std(list_winnings)/len(list_winnings)**.5
    capital = sum([x[1] for x in winnings_list if x[1]>0])
    print(winnings/capital,winnings,capital ,winnings-capital)
    print (winnings_stE*len(list_winnings),np.mean(net_profit),np.std(net_profit))
    print(np.mean([x<=0 for x in  net_profit]), np.mean([x for x in  net_profit if x >0]) ,max_payoff)
    import matplotlib.pyplot as plt
    plt.plot(np.convolve([x[0]-x[1] for x in winnings_list],[1./90,]*90,'valid'))
    plt.plot(np.convolve([x[0]-x[1] for x in winnings_list if x[0]>=0],[1./90,]*90,'valid'))
    plt.show()
    #print ((np.sum(df[var_used]*df['play'])-np.sum(df['play']))/np.sum(df['play']))           




    
```
